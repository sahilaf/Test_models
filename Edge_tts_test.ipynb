```python
import edge_tts
import asyncio
import IPython.display as ipd
import time
import tempfile
```

# --- Define Bangla TTS function ---

```python
async def bangla_tts(text, voice="bn-BD-NabanitaNeural", rate=10, pitch=0):
    """
    Convert Bangla text to speech using Edge-TTS in Colab.
    """
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    rate_str = f"{rate:+d}%"
    pitch_str = f"{pitch:+d}Hz"
    output_path = f"/content/bangla_tts_{timestamp}.mp3"

    communicate = edge_tts.Communicate(text, voice, rate=rate_str, pitch=pitch_str)
    await communicate.save(output_path)
    print(f"✅ Saved audio file: {output_path}")
    return output_path


# --- Wrapper for Colab-safe execution ---
def run_async(coro):
    """
    Run async functions safely inside Google Colab.
    """
    return asyncio.get_event_loop().create_task(coro)


# --- Example usage ---
text = "একটি এলএলএম কি?"
voice = "bn-BD-PradeepNeural"  # other options: bn-IN-TanishaaNeural, bn-IN-BashkarNeural, bn-BD-PradeepNeural

task = run_async(bangla_tts(text, voice))

# Wait for it to finish and get result
output_file = await task

# --- Play the audio directly ---
ipd.Audio(output_file)
```
